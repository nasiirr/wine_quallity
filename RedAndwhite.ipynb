```python

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import accuracy_score , classification_report
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve
from sklearn.preprocessing import StandardScaler
import gradio as gr
```

```python
wine_ds=pd.read_csv("redAndWhite.csv")
print(wine_ds.shape)
wine_ds.isnull().sum()
wine_ds.describe
```

```python
wine_ds['type'] = wine_ds['type'].map({'red': 0, 'white': 1})
```

```python
# get data with wine quality greater than 4 and less than 8
#wine_ds = wine_ds[(wine_ds['quality'] > 4) & (wine_ds['quality'] < 8)]

# reset index and drop the old one
wine_dss = wine_ds  
#.reset_index(drop=True)

wine_dss.to_csv('data.csv', index=False)
```

```python
X=wine_dss.drop('quality',axis=1)
Y=wine_dss['quality']
Y.to_csv('data.csv', index=False)
#train test  split
X_train , X_test , Y_train ,Y_test= train_test_split(X , Y , test_size=0.2 , random_state=2)
```

```python
print(Y.shape , Y_train.shape, Y_test.shape)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
X_test_preedection=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_preedection,Y_test)
print('accuracy :', test_data_accuracy)
```

```python
def predict_wine_quality_array(input_array):
    # Convert the comma-separated string to a list of numbers
    input_values = [float(val) for val in input_array.split(',')] 
    input_data_reshape = np.array(input_values).reshape(1, -1)
    prediction = model.predict(input_data_reshape)
    if prediction[0] == 0:
        pred = 'wine quality is bad'
    else:
        pred = 'wine quality is good'
    return pred

def predict_wine_quality_individual(type, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                                     free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol):
    # Convert the 'type' input to an integer (0 for Red, 1 for White)
    if type == 'Red':
        type = 0
    else:
        type = 1

    input_data = (type, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                  free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol)
    input_data_as_nparray = np.asarray(input_data)
    input_data_reshape = input_data_as_nparray.reshape(1, -1)

    prediction = model.predict(input_data_reshape)
    if prediction[0] == 0:
        pred = 'wine quality is bad'
    else:
        pred = 'wine quality is good'
    return pred , prediction[0]

# Mean values
mean_values = [6.9, 0.21, 0.33, 1.8, 0.034, 48, 136, 0.9899, 3.25, 0.41, 12.6]

# Create Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## Wine Quality Predictor")
    gr.Markdown("Select input method:")

    with gr.Row():
        button_array = gr.Button("Input Array")
        button_individual = gr.Button("Individual Input")

    with gr.Row():
        input_array = gr.Textbox(label="Enter values as comma-separated list (e.g., 0,7,0.2,0.3,...)",
                                 value=",".join(str(val) for val in mean_values))
        
        type_radio = gr.Radio(['Red', 'White'], label="Type")
        fixed_acidity = gr.Number(label="Fixed Acidity", value=mean_values[0])
        volatile_acidity = gr.Number(label="Volatile Acidity", value=mean_values[1])
        citric_acid = gr.Number(label="Citric Acid", value=mean_values[2])
        residual_sugar = gr.Number(label="Residual Sugar", value=mean_values[3])
        chlorides = gr.Number(label="Chlorides", value=mean_values[4])
        free_sulfur_dioxide = gr.Number(label="Free Sulfur Dioxide", value=mean_values[5])
        total_sulfur_dioxide = gr.Number(label="Total Sulfur Dioxide", value=mean_values[6])
        density = gr.Number(label="Density", value=mean_values[7])
        pH = gr.Number(label="pH", value=mean_values[8])
        sulphates = gr.Number(label="Sulphates", value=mean_values[9])
        alcohol = gr.Number(label="Alcohol", value=mean_values[10])

    output = [gr.Textbox(label="Prediction"),prediction]

    # Connect buttons to functions
    button_array.click(fn=predict_wine_quality_array, inputs=input_array, outputs=output)

    button_individual.click(fn=predict_wine_quality_individual,
                          inputs=[type_radio, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                                 free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol],
                          outputs=output)

    # Correct visibility toggle using .change events
    def show_array_input():
        input_array.visible = True
        type_radio.visible = False
        fixed_acidity.visible = False
        volatile_acidity.visible = False
        citric_acid.visible = False
        residual_sugar.visible = False
        chlorides.visible = False
        free_sulfur_dioxide.visible = False
        total_sulfur_dioxide.visible = False
        density.visible = False
        pH.visible = False
        sulphates.visible = False
        alcohol.visible = False
        return input_array, type_radio, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides, \
               free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol

    def show_individual_input():
        input_array.visible = False
        type_radio.visible = True
        fixed_acidity.visible = True
        volatile_acidity.visible = True
        citric_acid.visible = True
        residual_sugar.visible = True
        chlorides.visible = True
        free_sulfur_dioxide.visible = True
        total_sulfur_dioxide.visible = True
        density.visible = True
        pH.visible = True
        sulphates.visible = True
        alcohol.visible = True
        prediction.visible= True
        return input_array, type_radio, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides, \
               free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol,

    # **Correctly connecting button clicks to visibility functions**
    button_array.click(
        fn=show_array_input,
        inputs=[],
        outputs=[input_array, type_radio, fixed_acidity, volatile_acidity, citric_acid,
                 residual_sugar, chlorides, free_sulfur_dioxide, total_sulfur_dioxide,
                 density, pH, sulphates, alcohol,]
    )

    button_individual.click(
        fn=show_individual_input,
        inputs=[],
        outputs=[input_array, type_radio, fixed_acidity, volatile_acidity, citric_acid,
                 residual_sugar, chlorides, free_sulfur_dioxide, total_sulfur_dioxide,
                 density, pH, sulphates, alcohol,]
    )

demo.launch()
```

```python
def predict_wine_quality_array(input_array):
    # Convert the comma-separated string to a list of numbers
    input_values = [float(val) for val in input_array.split(',')] 
    input_data_reshape = np.array(input_values).reshape(1, -1)
    prediction = model.predict(input_data_reshape)
    if prediction[0] == 0:
        pred = 'wine quality is bad'
    else:
        pred = 'wine quality is good'
    return pred

def predict_wine_quality_individual(type, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                                     free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol):
    # Convert the 'type' input to an integer (0 for Red, 1 for White)
    if type == 'Red':
        type = 0
    else:
        type = 1

    input_data = (type, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
                  free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohol)
    input_data_as_nparray = np.asarray(input_data)
    input_data_reshape = input_data_as_nparray.reshape(1, -1)

    prediction = model.predict(input_data_reshape)
    if prediction[0] == 0:
        pred = 'wine quality is bad'
    else:
        pred = 'wine quality is good'
    return pred

# Mean values
mean_values = [6.9, 0.21, 0.33, 1.8, 0.034, 48, 136, 0.9899, 3.25, 0.41, 12.6]

# Create Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## Wine Quality Predictor")
    gr.Markdown("Select input method:")

    with gr.Row():
        button_array = gr.Button("Input Array")
        button_individual = gr.Button("Individual Input")

    # Create a container for the inputs
    with gr.Row(visible=True) as input_container:  # Make container visible by default
        input_array_textbox = gr.Textbox(label="Enter values as comma-separated list (e.g., 0,7,0.2,0.3,...)",
                                 value=",".join(str(val) for val in mean_values))

        # Individual inputs (visible by default)
        with gr.Group(visible=True) as individual_inputs:  # Make group visible by default
            type_radio_individual = gr.Radio(['Red', 'White'], label="Type")
            fixed_acidity_individual = gr.Number(label="Fixed Acidity", value=mean_values[0])
            volatile_acidity_individual = gr.Number(label="Volatile Acidity", value=mean_values[1])
            citric_acid_individual = gr.Number(label="Citric Acid", value=mean_values[2])
            residual_sugar_individual = gr.Number(label="Residual Sugar", value=mean_values[3])
            chlorides_individual = gr.Number(label="Chlorides", value=mean_values[4])
            free_sulfur_dioxide_individual = gr.Number(label="Free Sulfur Dioxide", value=mean_values[5])
            total_sulfur_dioxide_individual = gr.Number(label="Total Sulfur Dioxide", value=mean_values[6])
            density_individual = gr.Number(label="Density", value=mean_values[7])
            pH_individual = gr.Number(label="pH", value=mean_values[8])
            sulphates_individual = gr.Number(label="Sulphates", value=mean_values[9])
            alcohol_individual = gr.Number(label="Alcohol", value=mean_values[10])

    output = gr.Textbox(label="Prediction")

    # Connect buttons to functions
    button_array.click(
        fn=predict_wine_quality_array,
        inputs=input_array_textbox,
        outputs=output
    )

    button_individual.click(
        fn=predict_wine_quality_individual,
        inputs=[type_radio_individual, fixed_acidity_individual, volatile_acidity_individual, citric_acid_individual, residual_sugar_individual, chlorides_individual,
                free_sulfur_dioxide_individual, total_sulfur_dioxide_individual, density_individual, pH_individual, sulphates_individual, alcohol_individual],
        outputs=output
    )

    # Correct visibility toggle using .change events
    def show_array_input():
        input_container.visible = True
        individual_inputs.visible = False
        return input_container, individual_inputs

    def show_individual_input():
        input_container.visible = False
        individual_inputs.visible = True
        return input_container, individual_inputs

    button_array.click(
        fn=show_array_input,
        inputs=[],
        outputs=[input_container, individual_inputs]
    )

    button_individual.click(
        fn=show_individual_input,
        inputs=[],
        outputs=[input_container, individual_inputs]
    )

demo.launch()
```
